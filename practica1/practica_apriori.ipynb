### 1.-Instalar bibliotecas

````python
import numpy as np
import matplotlib.pyplot as plt  
import pandas as pd  
from itertools import combinations
````

### 2.-Importar archivo 


````python
canciones= np.load('spotify.npy',allow_pickle=True).tolist()
````

```hallamos las frecuencias de cada cancion en las playlist```

````python
frec = {}
i=0
j=0
for i in range(len(canciones)):
  for j in range(len(canciones[i])):
    palabra=canciones[i][j]
    if palabra in frec:
      frec[palabra] += 1
    else:
      frec[palabra] = 1
    j=j+1
  i=i+1
songs_1er= frec
songs_1er
````

```seguidamente analizamos los diciconarion para ver si pasan el umbral realizmaos una combinacion de 2 en 2```

````python
arr=[]
MINsup=0.020
#verificar umbreal
for key, value in songs_1er.items():
  aux = value/10000
  if(aux>= MINsup):
    arr.append(key)
#combiar
songs = list(combinations( arr, 2 ))
print(songs)
````

```cuenta las combinaciones obtenidas , en el conjunto inicial de datos ```

````python
arr_2da={}
for i in range(len(canciones)):
  rep=0
  for j in range(len(songs)):
    if(set(canciones[i]).intersection(list(songs[j]))):
      rep=rep+1
    arr_2da[songs[j]]=rep
````

```seguidamente analizamos los diccionario para ver si pasan el umbral```

````python
arr1=[]
MINsup=0.010
#ver si pasa umbral
for key, value in arr_2da.items():
  aux = value/10000
  if(aux>= MINsup):
    arr1.append(key)
````
